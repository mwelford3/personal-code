```python
# Import required libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import joined data.
data = pd.read_csv('stage1joined_data.csv')
```

```python
# Define features to have correlations calculated.
features = ['Start Year', 'Completion Year', 'Blog mentions', 'Policy mentions', 'Patent mentions', 'Weibo mentions', 'Wikipedia mentions','Q&A mentions', 'Video mentions',  
            'Altmetric Title Length','News mentions','Twitter mentions','Facebook mentions','Google+ mentions','Total Mentions','Numeric Phase', 'Number of Phases', 'Number of Sponsors', 'Altmetric Attention Score']
```

```python
# Subset the data.
data_sub = data[features]
```

```python
# Print the data.
data_sub
```

```python
# Extract the correlations.
corr = data_sub.corr()
```

```python
# Print the description of the joined data.
data_sub.describe()
```

```python
# Print the variances of the features.
data_sub.var()
```

```python
# Show the correlation headmap for the features.
fig = sns.heatmap(corr, annot=True, fmt='.2f')
plt.savefig('numeric_features_heatmap.png')
```
