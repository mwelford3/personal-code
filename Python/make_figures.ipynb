```python
# Import required libraries.
import pandas as pd
import matplotlib.pyplot as plt
```

### Model Perfromances on Numeric Features

```python
# Import Numeric Model Results
numeric_model_results = pd.read_csv('NumericResults.csv')
numeric_model_results.fillna('N/A', inplace=True)
numeric_model_results_table = numeric_model_results.set_index('Classifier')
numeric_model_results_table
```

```python
# Numeric Training Accuracy
numeric_model_results.plot(x='Classifier', y='Training Accuracy', kind='bar')
```

```python
# Numeric Test Accuracy
numeric_model_results.plot(x='Classifier', y='Test Accuracy', kind='bar')
```

```python
numeric_model_results = numeric_model_results[numeric_model_results['Classifier'] != 'XGBoost using Dmatrix']
# Numeric Model Group 0 Precision
numeric_model_results.plot(x='Classifier', y='Group 0 Precision', kind='bar')
```

```python
# Numeric Model Weighted Avg. Precision
numeric_model_results.plot(x='Classifier', y='Weighted Avg. Precision', kind='bar')
plt.show()
```

```python
# Numeric Model Group 0 Recall
numeric_model_results.plot(x='Classifier', y='Group 0 Recall', kind='bar')
plt.show()
```

```python
# Numeric Model Weighted Avg. Recall
numeric_model_results.plot(x='Classifier', y='Weighted Avg. Recall', kind='bar')
plt.show()
```

```python
# Numeric Model Weighted Avg F1 Score
numeric_model_results.plot(x='Classifier', y='Weighted Avg. F1 Score', kind='bar')
plt.show()
```

```python
# Numeric Model Specificity of 0 group
numeric_model_results.plot(x='Classifier', y='Specificity of Group 0', kind='bar')
plt.show()
```

```python
# Numeric Model Cross Validation Accuracy
numeric_model_results.plot(x='Classifier', y='CV Accuracy', kind='bar')
plt.show()
```
