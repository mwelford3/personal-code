```python
import pandas as pd
import altair as alt
```

```python
new_data = pd.read_csv('../data/Long-Term Care Facility Data 4082022.csv')
```

```python
new_data.head()
```

```python
new_data['Total Deaths'] = new_data['Staff Total COVID-19 Deaths'] + new_data['Resident Total COVID-19 Deaths']
new_data.sort_values('Total Deaths', ascending=False).head(10)
```

```python
alt.Chart(new_data.sort_values('Total Deaths', ascending=False).head(10)).mark_bar().encode(
    y='Facility Name',
    x='Total Deaths')
```
