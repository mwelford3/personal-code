```python
import pandas as pd
import altair as alt
```

```python
long_term_care_data = pd.read_csv('../data/Long-Term Care Facility Data New.csv')
long_term_care_data.head()
```

```python
has_resident_vaccination_rate = long_term_care_data['Resident Vaccination Rate'] != 'No Value'
has_staff_vaccination_rate = long_term_care_data['Staff Vaccination Rate'] != 'No Value'

all_data = long_term_care_data[has_resident_vaccination_rate & has_staff_vaccination_rate]


all_data['Resident Vaccination Rate'] = all_data['Resident Vaccination Rate'].str.rstrip('%').astype(float)
all_data['Staff Vaccination Rate'] = all_data['Staff Vaccination Rate'].str.rstrip('%').astype(float)

all_data
```

```python
northern_counties_5 = ['Stephenson', 'Winnebago', 'Boone', 'Ogle','DeKalb']
```

```python
northern_data = all_data[all_data['County'].isin(northern_counties_5)]
```

```python
data_plot2 = northern_data.sort_values('Staff Vaccination Rate', ascending=False).head(10)
data_plot2
```

```python
name2=['Oak Crest','Ascension Saint Anne Place', 'Rock River Health Care', 'Symphony Northwoods', 'Ascension Saint Joseph Village', 'River Bluff Nursing Home','Stephenson Nursing Center','Mercyhealth Javon Bea Hospital','Aperion Care Dekalb','Forest City Rehab']

data_plot2['clean_name']=name2
```

```python
# Create the chart.
chart = alt.Chart(data_plot2).mark_bar().encode(
    y=alt.Y('clean_name', title="Facility Name",sort='-x'),
    x=alt.X('Staff Vaccination Rate',
           title="Percent Completely Vaccinated Staff (%)"),
    color=alt.Color('Region',scale=alt.Scale(range=['#ce7067']))).properties(padding={'top':10, 'bottom':10, 'left':25, "right":30},title={"text":'Top 10 Facilities by Percentage Vaccinated Staff in The Northern 5 Counties (as of 5/22/2022)', "subtitle":'Labeled by Region'})
# Configure the fonts.
chart.configure_title(font='Century Gothic',
                     fontSize = 14).configure_axisX(
                        labelFont='Century Gothic').configure_axisY(labelFont='Century Gothic', labelLimit=300, titleX=-250)
```
